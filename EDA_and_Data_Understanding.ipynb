# Get To Know Data and Basic Statistics

**Get To Know Data**
---

In a dataset, there are columns and rows
*   Columns is the side or horizontal part. Or it can also be said to be a type of data (e.g Age, Name, Height)
*   Rows is the downward or vertical part, or can also be said to be the data (e.g 20, Asya, 168)

There is also a data types
*   **Numerical** is a data that is calculated or operated mathematically (e.g 20, 89,0, 50000, etc). There is an **Integer** (e.g 10, 21, 33, 90, etc) and **Float** (e.g 5.10, 10.1, 6.7, etc)

*  **Categorical** is a categori data or label data, not a numbers that we can coount mathematically. There is a **Nominal Data**, unordered data (e.g gender or city) and **Ordinal Data**, ordered data (e.g education level, such as high school > middle school > elementary school)

*  **Datetime** is a data that related to time (e.g 2023-12-09, 2025/10/09 07:00, 19/20/2024, etc)



**Basic Statistics**
---

*   **Mean**
---

Mean is the average value of a data.


Mean.png

(e.g (10 + 20 + 30)/3 = 20)

Mean = 20


*   **Median**
---

Median is the middle value.

If number of data is odd → middle value

If number of data is even → average of two middle values

(e.g Data = [10, 20, 30, 40] → Median = (20 + 30) / 2 = 25)


*   **Mode**
---
Mode is values of data that appears most often


*   **Variance**
---

Is how spread out the data is


population-variance.png


* **Standard Deviation**
---

STD is avarage distance from the mean. It is small or large.

STD formula is a square root of Variance


* **Skewness**
---

Tells which side the data tail is longer:


Right (positive skew): tail to the right


Left (negative skew): tail to the left


0 = symmetric (normal)


Kewwness.png


* **Precentiles**
---

Shows position of a value in the data.


25th percentile = 25% of data below it, 50th = median, 75th = 75% of data below it


precentile.jpg

* **Correlations and Covarience**
---
**Correlations** is measures the strength and direction of the relationship between two variables.
Correlation is a standardized version of covariance, the result is always between -1 and 1.

Use
```
# print(df.corr())
```
to see the correlations


**Covariance** is measuring the direction of the relationship between two numerical variables.
You could say, if one increases, does the other increase or decrease?

Covariance values ​​are not standardized, their values ​​depend on the scale of the data.

Use
```
# print(df.cov())
```
to see the covariance


**Tools**
---

* **Pandas**
---
Library for managing and analyzing data in tabular form (DataFrame).


Functions:
- Read data (.csv, .xlsx, .json, etc.)

A basic library for mathematical calculations and numeric arrays.

Functions:
- Fast mathematical operations (averages, square roots, logs, etc.)

* **Seaborn**
---
A visualization library based on Matplotlib, but with a more beautiful and automated interface.

Functions:

- Automatically elegant colors and styles

````python
plt.figure(figsize=(8,5))
sns.histplot(df['votes'], bins=30, kde=True)
plt.title('Distribution of Votes')
plt.xlabel('Votes')
plt.ylabel('Frequency')
plt.show()
````

````python
plt.figure(figsize=(8,5))
sns.histplot(np.log1p(df['votes']), bins=30, kde=True)
plt.title('Log-transformed Distribution of Votes')
plt.xlabel('log(Votes + 1)')
plt.ylabel('Frequency')
plt.show()
````

````python
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
plt.figure(figsize=(12, 8))
for i, col in enumerate(numeric_cols, 1):
    plt.subplot(2, 2, i)
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'{col} (Skew = {df[col].skew():.2f})')
plt.tight_layout()
plt.show()
````

````python
votes_log = np.log1p(df['votes'])

mean_val = votes_log.mean()
median_val = votes_log.median()

plt.figure(figsize=(8,5))
sns.histplot(votes_log, bins=30, kde=True)

plt.axvline(mean_val, color='red', linestyle='-', linewidth=1, label=f'Mean = {mean_val:.2f}')
plt.axvline(median_val, color='green', linestyle='-', linewidth=1, label=f'Median = {median_val:.2f}')

plt.title('Log-transformed Distribution of Votes')
plt.xlabel('log(Votes + 1)')
plt.ylabel('Frequency')
plt.legend()
plt.show()
````

````python
plt.figure(figsize=(8,5))
sns.histplot(df['users_rating'], bins=30, kde=True, color='gray')

mean_val = df['users_rating'].mean()
median_val = df['users_rating'].median()

plt.axvline(mean_val, color='red', linestyle='-', linewidth=1, label=f'Mean = {mean_val:.2f}')
plt.axvline(median_val, color='green', linestyle='-', linewidth=1, label=f'Median = {median_val:.2f}')

plt.title('Distribution of Users Rating')
plt.xlabel('Users Rating')
plt.ylabel('Frequency')
plt.legend()

plt.show()
````

````python
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['runtime'])
plt.title('Spread and Outliers of Runtime')
plt.xlabel('Runtime')
plt.show()
````

````python
plt.figure(figsize=(8, 5))
sns.heatmap(df[['runtime', 'users_rating']].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation: Runtime and Users Rating')
plt.show()
````

````python
plt.figure(figsize=(6,4))
sns.scatterplot(x='runtime', y='users_rating', data=df)
plt.title('Runtime vs User Rating')
plt.show()
````

# Data Celeaning and Missing Values

Data cleaning is the stage for cleaning and repairing raw data so that it is ready to be used for analysis and models.

**Missing Values**

Missing values ​​are usually marked with NaN (Not a Number) or None or ? even -

**Type of Missing Value**
* **MCAR** (Missing Completely At Random) is a type of missing values ​​with missing values ​​that are truly random and do not depend on other variables The solution to MCAR is to delete rows or columns without major problems

* **MAR** (Missing At Random) is a missing value that depends on another variable, such as salary with a missing value at a certain age, here it can be said that age affects the missing values. The solution to this is to fill in the missing values ​​with the average salary for the same age.

* **MNAR** (Missing Not At Random) is missing values ​​are those that depend on the value itself. For example, people with high salaries don't want their salaries recorded, resulting in missing values. This problem is difficult to solve, we'll explore this in machine learning.
